```python
%matplotlib nbagg
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
fig, ax = plt.subplots() #utiliser plt qu'une suele fois, un seul dessin, un objet 
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100) #gÃ©nÃ¨re 100 point en min et max donnÃ©

ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6)

#ax.plot(x, np.full(len(x), 0.2),
#      'b-', lw=1)

fig.show()
```

```python
import numpy
import math 
mean=numpy.mean(df['Log2 Corrected Abundance Ratio'])
ec=numpy.std(df['Log2 Corrected Abundance Ratio'])*numpy.std(df['Log2 Corrected Abundance Ratio'])
n=len(df['Log2 Corrected Abundance Ratio'])
ec_corrigÃ©e=math.sqrt((n/(n+1))*ec)
print(ec_corrigÃ©e)

fig, ax = plt.subplots() #utiliser plt qu'une suele fois, un seul dessin, un objet 
#x = df['Log2 Corrected Abundance Ratio']
x = df['Log2 Corrected Abundance Ratio']
hist = ax.hist(x, 50)
dx=hist[1][1] - hist[1][0]
scale = len(x)*dx

ax.set_xlabel('Log2 Corrected Abundance Ratio')
ax.set_ylabel('protein count')

x = np.linspace(-4, 1.5, 100)
ax.plot(x, norm.pdf(x,mean,ec_corrigÃ©e)*scale)


fig.savefig("hist_norm.png")
```
